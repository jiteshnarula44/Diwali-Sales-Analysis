## Diwali Sales Analysis


### Importing libraries for cleaning and visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

#### Importing Dataset

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\Diwali Sales Data.csv", encoding = 'unicode escape')
```

```python
pd.set_option('display.min_rows',None)
```

#### Dropping Unnecessary Columns

```python
df = df.drop(['Status','unnamed1'],axis = 1)
```

```python
pd.isnull(df).sum()
```

```python
#Handling Null Values
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int32')
```

### Exploratory Data Analysis

#### Gender

```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

#### Age

```python
sns.countplot(data = df, x ='Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x= 'Age Group', y = 'Amount', data = sales_age)
```

#### Occupation

```python
sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```
